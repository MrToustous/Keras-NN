```python
def get_logreturns(data):
    return pd.DataFrame(np.log(data).diff(), index = data.index[1:]) 

def clean_data(data):
    return data.dropna()

def data_window(data, start = None, end = None):
    if not start:
        start_date = data.index[0]
    else:
        start_date = pd.to_datetime(start, dayfirst=True)
        
    if not end:
        end_date = data.index[-1]
    else:
        end_date = pd.to_datetime(end,  dayfirst=True)
        
    return data.loc[pd.to_datetime(start_date, dayfirst = True):pd.to_datetime(end_date, dayfirst = True)]

def Garch_CV_metrics(data, n_folds = 10, **params):
    data = get_logreturns(data['Adjusted Close'])
    results_aic = []
    results_bic = []
    results_loglik = []
    first_date = pd.to_datetime(data.index[0])
    last_date = pd.to_datetime(data.index[-1])
    time_span = (last_date - first_date).days
    splits =  [first_date + pd.Timedelta(days = int(np.random.rand(1)*time_span)) for _ in range(n_folds)]
    for x in splits:
        train_set = data_window(data, first_date, x)
        model = pf.GARCH(train_set, **params)
        x = model.fit()
        results_aic.append(x.aic)
        results_bic.append(x.bic)
        results_loglik.append(x.loglik)
    return np.mean(results_loglik), np.mean(results_aic), np.mean(results_bic)


def make_grid(parameters,default = {}):
    """Recursive function which turns a dictionary of list of parameters into a list of dictionary of parameters usable for a grid search"""
    grid = []
    for parameter in parameters[0][1]:
        grid_param = default.copy()
        grid_param[parameters[0][0]]=parameter
        if parameters[0][0] == parameters[-1][0]:
            grid += [grid_param]
        else:
            grid += make_grid(parameters[1:],default = grid_param)
    return grid
```

```python
%%Time

parameters = {'p': [2, 4, 6, 8], 
              'q':[2, 4, 6, 8]}
parameters = sorted(list(parameters.items()),key = lambda x: len(x[1]),reverse = True)
params = make_grid(parameters)
results = {'CV_loglik': [], 
           'CV_aic': [], 
           'CV_bic': [], 
           'Parameters': []}
                   
for param in params:
    CV_loglik, CV_aic, CV_bic = Garch_CV_metrics(google, **param)
    results['CV_loglik'].append(CV_loglik)
    results['CV_aic'].append(CV_aic)
    results['CV_bic'].append(CV_bic)
    results['Parameters'].append(param)
    
results = pd.DataFrame(results)
```
